# Analysis

## Cycle size VS minimum number of cards

```python
# Get maximum cycle size
cycles_size_col = df["CYCLES_MULTIPLE_OF_NB_CARDS_LIST"].to_list()
cycles_size_col_without_nan = [ i for i in cycles_size_col if isinstance(i, list) ]
n = int(max([float(item) for sublist in cycles_size_col_without_nan for item in sublist]))
# Get min nb cards to get this cycle multiple
result = {}
for cycle_multiple in range(1, n+1):
    nb_cards = df.loc[df["CYCLES_MULTIPLE_OF_NB_CARDS_LIST"].str.contains("%s.0" %cycle_multiple, regex=False, na=False), "NB_CARDS"].min()
    result[cycle_multiple] = nb_cards
```

```python
for key, value in result.items():
    print(key, value)
```

- for one cycle size, this size will appear again for each multiple of the minimun number of cards where it appears first. Example : for the cycle size 12, it appears for the number of cards 14, 28, 42, 56...
